```python
import pandas as pd
import zipfile

# 1. Unzip the file
# Use the absolute path to the zip file as indicated in the kernel state
with zipfile.ZipFile('/content/Metro_Interstate_Traffic_Volume.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/') # Extract to the /content directory

# 2. Load the CSV (check the sidebar to make sure the name matches exactly)
# The CSV should now be in /content, so we can directly read it
df = pd.read_csv('/content/Metro_Interstate_Traffic_Volume.csv')

# 3. Quick check
print(df.head())
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 1. Scale the traffic volume between 0 and 1
scaler = MinMaxScaler()
scaled_traffic = scaler.fit_transform(df[['traffic_volume']])

# 2. Create sequences (Windowing)
def create_sequences(data, window_size):
    X = []
    y = []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

# We will look at the last 24 hours to predict the next 1 hour
window_size = 24
X, y = create_sequences(scaled_traffic, window_size)

print(f"Shape of Input (X): {X.shape}") # (Number of samples, 24 hours, 1 feature)
print(f"Shape of Target (y): {y.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 1. Define the architecture
model = Sequential([
    LSTM(50, activation='relu', input_shape=(window_size, 1)),
    Dense(1) # The single predicted traffic value
])

# 2. Compile the model
model.compile(optimizer='adam', loss='mse')

# 3. Train the model (This might take a minute)
# We use a small part of the data (10 epochs) just to test it
model.fit(X, y, epochs=10, batch_size=32)
```

```python
# 1. Ask the AI to predict traffic for the whole dataset
predictions = model.predict(X)

# 2. Undo the scaling (Turn decimals back into car counts)
# We undo it for both the predictions and the actual 'y' values
predictions_rescaled = scaler.inverse_transform(predictions)
y_rescaled = scaler.inverse_transform(y.reshape(-1, 1))

print("First 5 Actual Traffic Volumes:", y_rescaled[:5].flatten())
print("First 5 AI Predictions:", predictions_rescaled[:5].flatten())
```

```python
import matplotlib.pyplot as plt

# Let's zoom into a specific 100-hour window to see the detail
plt.figure(figsize=(15, 6))
plt.plot(y_rescaled[100:200], label='Actual Traffic (The Truth)', color='blue', linewidth=2)
plt.plot(predictions_rescaled[100:200], label='AI Prediction', color='red', linestyle='--', linewidth=2)

plt.title('Traffic Jam Prediction vs Reality')
plt.xlabel('Time (Hours)')
plt.ylabel('Number of Cars')
plt.legend()
plt.show()
```

```python
# 1. Grab the last 24 hours from your data
last_24_hours = scaled_traffic[-24:]

# 2. Reshape it so the AI understands it's 1 sequence
last_24_hours = last_24_hours.reshape(1, 24, 1)

# 3. Predict!
next_hour_prediction_scaled = model.predict(last_24_hours)

# 4. Rescale
next_hour_cars = scaler.inverse_transform(next_hour_prediction_scaled)

print(f"The AI predicts there will be {int(next_hour_cars[0][0])} cars on the road in the next hour.")
```

```python
# 1. Grab the very last 24-hour window from the dataset
recent_history = scaled_traffic[-24:]

# 2. Reshape it so the AI knows it's 1 single sequence
recent_history = recent_history.reshape(1, 24, 1)

# 3. Ask the AI for the prediction
future_scaled = model.predict(recent_history)

# 4. Turn the decimal back into a car count
future_cars = scaler.inverse_transform(future_scaled)

# 5. Output the result
predicted_count = int(future_cars[0][0])
print(f"Prediction for the next hour: {predicted_count} cars.")
```

```python
# Define a threshold (e.g., if more than 3500 cars, it's a jam)
threshold = 3500

if predicted_count > threshold:
    print("Warning: Heavy Traffic/Jam predicted! Consider an alternate route.")
else:
    print("Traffic is moving smoothly.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

# 1. Compare all predictions against the actual truth from the CSV
mae = mean_absolute_error(y_rescaled, predictions_rescaled)
rmse = math.sqrt(mean_squared_error(y_rescaled, predictions_rescaled))

# 2. Print the results
print(f"Mean Absolute Error: {int(mae)} cars")
print(f"Root Mean Squared Error: {int(rmse)} cars")

# 3. Calculate Accuracy Percentage (rough estimate)
avg_traffic = df['traffic_volume'].mean()
accuracy = 100 - (mae / avg_traffic * 100)
print(f"Approximate Accuracy: {accuracy:.2f}%")
```

```python
# 1. Select more columns this time
features = ['traffic_volume', 'temp', 'rain_1h']
data_multivariate = df[features].values

# 2. Scale all features (each column gets its own 0-1 scaling)
scaler_multi = MinMaxScaler()
scaled_data_multi = scaler_multi.fit_transform(data_multivariate)

# 3. Create sequences again (using the same window_size of 24)
X_multi, y_multi = create_sequences(scaled_data_multi, window_size)

# Note: y_multi is currently all 3 features, but we only want to predict traffic_volume (index 0)
y_multi = y_multi[:, 0]

print(f"New Input Shape: {X_multi.shape}")
# It should show (samples, 24, 3) -> The '3' means Traffic, Temp, and Rain!
```

```python
model_multi = Sequential([
    LSTM(64, activation='relu', input_shape=(window_size, 3), return_sequences=True),
    LSTM(32, activation='relu'),
    Dense(1)
])

model_multi.compile(optimizer='adam', loss='mse')

# Let's train for 15 epochs this time since the data is more complex
model_multi.fit(X_multi, y_multi, epochs=15, batch_size=32)
```
